```python
import pandas as pd 
import matplotlib.pyplot as plt 
from wordcloud import WordCloud, STOPWORDS
```

```python
f = open("../tai_nguyen_cung_cap/fb_data.txt", "r", encoding='utf-8') 
data = f.readline()
f.close()
```

```python
wc = WordCloud(background_color='black', max_words=1000, stopwords=STOPWORDS )
# generate the word cloud 
wc.generate(data)
```

```python
# display the word clouds 
plt.figure(figsize=(10, 12)) 
plt.imshow(wc, interpolation='bilinear')
plt.axis('off') 
plt.show()
```

```python
import numpy as np 
from PIL import Image
```

```python
wc_mask = np.array(Image.open('database.png'))
```

```python
# instantiate a word cloud object 
wc1 = WordCloud(background_color='black', max_words=1000, mask=wc_mask, stopwords=STOPWORDS) 
# generate the word cloud 
wc1.generate(data) 
# display the word cloud 
plt.figure(figsize=(10, 12)) 
plt.imshow(wc1, interpolation='bilinear') 
plt.axis('off') 
plt.show()
```
